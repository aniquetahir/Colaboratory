```
from robobrowser import RoboBrowser
browser = RoboBrowser(history=False)
```

```
# Go to cirnopedia servants page
browser.open('http://fate-go.cirnopedia.org/servant_all.php')
```

```
servants_table = browser.select('table#rounded-corner')[0]
```

```
servant_rows = servants_table.select('tr')
```

```
headers = [x.text for x in servant_rows[0].select('th')]
```

```
print(headers)
```

```
servants = []
for servant_div in servant_rows[2:]:
  divs = servant_div.select('td')
  servant_dict = {}
  final_dict = {}
  for i,div in enumerate(divs):
    servant_dict[headers[i]] = div
  final_dict['ID'] = int(servant_dict['ID'].text)
  final_dict['★'] = servant_dict['★'].text.strip()
  final_dict['Name'] = servant_dict['Name'].get('sorttable_customkey')
  final_dict['Class'] = servant_dict['Class'].text.strip()
  final_dict['Cost'] = int(servant_dict['Cost'].text)
  final_dict['BaseHP'] = int(servant_dict['BaseHP'].text.replace(',',''))
  final_dict['BaseATK'] = int(servant_dict['BaseATK'].text.replace(',',''))
  final_dict['MaxATK'] = int(servant_dict['MaxATK'].text.replace(',',''))
  final_dict['MaxHP'] = int(servant_dict['MaxHP'].text.replace(',',''))
  servants.append(final_dict)
```
